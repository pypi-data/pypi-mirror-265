## Polygons

```python
%%turtle 200

def poly(sides, size):
    turn = 360 / sides
    for _ in range(sides):
        fd(size)
        lt(turn)

hide()

with no_pen():
    bk(20)
    lt(90)
    bk(75)
    rt(90)
    
for n in range(3, 13):
    poly(n, 40)
```

```python
%%turtle 240
hide()
with no_pen():
    bk(10)
    lt(90)
    bk(115)
    rt(90)

for n in range(3, 37):
    poly(n, 20)
```

## Introduction to Procedures

Pages 10-14.

```python
%%turtle

def box():
    fd(60)
    rt(90)
    fd(30)
    rt(90)
    fd(60)
    rt(90)
    fd(30)

box()
```

```python
def pinwheel():  # PT-BR: catavento
    for _ in range(4):
        box()
```

```python
pinwheel()
```

```python
%%turtle

def repeat(times, fn):
    for _ in range(times):
        fn()

repeat(4, box)
    
```

```python
%%turtle

for _ in range(8):
    rt(45)
    box()
```

### Ramalho tripping

```python
%%turtle

def twist(times, angle, fn):
    angle = 360 / 4 / times
    for _ in range(times):
        lt(angle)
        fn()

def square():
    fd(80)
    lt(90)

with no_pen():
    lt(45)
    bk(60)
    
twist(15, 5, square)
```

```python
%%turtle

twist(30, 5, box)
```

```python
%%turtle

def repeat(times, fn, *args):
    for _ in range(times):
        fn(*args)

def pinwheel_rotated(angle):
    lt(angle)
    pinwheel()

repeat(6, pinwheel_rotated, 15)
```

## Procedures with inputs

```python
%%turtle

def rectangle(height, length):
    fd(height)
    rt(90)
    fd(length)
    rt(90)
    fd(height)
    rt(90)
    fd(length)
    rt(90)

rectangle(100, 10)
```

```python
make_turtle()     

def flag(height):
    fd(height)
    rectangle(height/2, height)
    bk(height)

lt(90)
flag(40)
    
```

```python
make_turtle()

def circle1():
    for _ in range(360):
        fd(1)
        rt(1)
        
with no_pen():
    lt(90)
    fd(55)
    rt(90)
    
circle1()
show_SVG()
```

```python
make_turtle()

def circle2(size):
    for _ in range(360):
        fd(size)
        rt(1)

with no_pen():
    lt(90)
    fd(55)
    rt(90)

circle2(1)
circle2(.5)
circle2(.1)
```

```python
from math import pi

make_turtle()

def circle(radius):
    circle2(radius * pi/180)


hide()

with no_pen():
    lt(90)
    fd(60)
    rt(90)
    
for r in range(10, 25, 5):
    circle(r)
```

```python
make_turtle()

hide()

with no_pen():
    lt(90)
    fd(60)
    rt(90)

with no_update():
    for r in range(10, 70, 5):
        circle(r)
```

```python
def arcright1(size, angle):
    for _ in range(angle):
        fd(size)
        rt(1)

def new_arcright(radius, angle):
    arcright1(radius*pi/180, angle)

draw_height = 200
make_turtle(height=draw_height)

jumpto(10, 10)
new_arcright(50, 60)

jumpto(70, 10)
set_heading(0)
new_arcright(100, 90)

jumpto(180, 10)
set_heading(0)
new_arcright(30, 180)

jumpto(220, 10)
set_heading(0)
new_arcright(300, 10)
```

```python
make_turtle(height=210)

def petal(size):
    new_arcright(size, 60)
    rt(120)
    new_arcright(size, 60)
    rt(120)

def flower(size):
    for _ in range(6):
        petal(size)
        rt(60)

flower(100)
```

```python
with no_update():
    flower(90)
    flower(80)
    flower(70)
```

```python
def arcleft1(size, angle):
    for _ in range(angle):
        fd(size)
        lt(1)

def new_arcleft(radius, angle):
    arcleft1(radius*pi/180, angle)

def ray(size):
    new_arcleft(size, 90)
    new_arcright(size, 90)
    new_arcleft(size, 90)
    new_arcright(size, 90)

def sun(size):
    for _ in range(9):
        ray(size)
        rt(160)

make_turtle(width=320, height=320)

jumpto(40, 250)

hide()
sun(50)
```
